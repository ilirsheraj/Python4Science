## f-Print and f-String

```python
num = 45.3
sng = "cellar doors"

print("I would like to have " + str(num) + " " + sng)
```

```python
# %g and %s has been inherited from fortran.
# %g stands for number and %s for string
print("I would like to have %g %s"%(num, sng))
```

```python
# %i is for integer
print("I would like to have %i %s"%(num, sng))
```

```python
# %f gives more options for formating
print("I would like to have %f %s"%(num, sng))
```

```python
# %.3f gives precision to 3 significant figures
print("I would like to have %.3f %s"%(num, sng))
```

```python
# number before f gives total number of spaces before the decimal
print("I would like to have %10.3f %s"%(num, sng))
```

```python
# Do math directly
print("I would like to have %g %s"%(num/5, sng))
```

```python
# f-strings
# f in front tells python to fill the curly brackets with the variables
print(f"I would like to have {num} {sng}.")
```

```python
print("I would like to have {n} {s}".format(n = num, s = sng))
```

```python
# My Solution
# Exercise
# Print out each letter of English alphabet
# a is the first letter of the english alphabet
# ...
# z is the 26th letter of the alphabet
import string
string.ascii_lowercase
for i in range(0,26):
    if i in [0,20]:
        print("{} is the {}st letter of the English Alphabet".format(string.ascii_lowercase[i], i+1))
    elif i in [1, 21]:
        print("{} is the {}nd letter of the English Alphabet".format(string.ascii_lowercase[i], i+1))
    elif i in [2, 22]:
        print("{} is the {}rd letter of the English Alphabet".format(string.ascii_lowercase[i], i+1))
    else:
        print("{} is the {}th letter of the English Alphabet".format(string.ascii_lowercase[i], i+1))
```

## Subplot Geometry

How to lay out multiple plots in a single area, called subplots

```python
plt.subplot(1,2,1) # 1 row, 2 columns, number of subplots
plt.plot(np.random.randn(10))

plt.subplot(1,2,2) # 1 row, 2 columns, number of subplots
plt.plot(np.random.randn(10))
plt.show()
```

```python
plt.subplot(2,1,1) # 1 row, 2 columns, number of subplots
plt.plot(np.random.randn(10))

plt.subplot(2,1,2) # 1 row, 2 columns, number of subplots
plt.plot(np.random.randn(10))
plt.show()
```

```python
# For simple cases
for i in range(9):
    plt.subplot(3,3,i+1)
    plt.plot(np.random.randn(7))
```

```python
# ax is a numpy array with three objects
fig,ax = plt.subplots(1,3, figsize = (15,3)) #rows, columns, width, height

x = np.arange(10)
ax[0].plot(x, x**2, "b") # first plot
ax[1].plot(x, np.sqrt(x), "r") # second plot
ax[2].plot(x,x, "g") # third plot
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(x,x, label = "x")
ax.plot(x, x**2, label = "x**2")
ax.plot(x, x**3, label = "x**3")
plt.legend()
# plt.xlabel("x") # error
ax.set_xlabel("x")
ax.set_ylabel("y = f(x)")
ax.set_title("Awesome Plot")
ax.set_xlim([-3,3])
# plt.gca() # access the axis
ax.set_aspect(1/ax.get_data_ratio())
ax.grid()
plt.show()
```

## Adding Annotations

```python
x = np.arange(-5,6)
y = x**2

fig,ax = plt.subplots()
ax.plot(x,y, "ko-", markerfacecolor = "w", markersize = 15)

# Specify an annotation
ax.annotate("Hello", (x[3],y[3]), fontsize = 15,arrowprops = dict(),
           xycoords="data", xytext = (0,10), horizontalalignment = "center")
plt.show()
```

```python
# Exercise
# Make a graph with annotation to find the minimum value in the function shown below
# wedge
x = np.arange(-5,6)
y = x**2 + np.cos(x)*10

minpnt = np.argmin(y)
txt = "min: (%g,%.2f)"%(x[minpnt], y[minpnt])

fig,ax = plt.subplots()
ax.plot(x,y, "ko-", markerfacecolor = "purple", markersize = 15)
ax.set_xlabel("x")
ax.set_ylabel("f(x) = x**2 + np.cos(x)*10")

ax.annotate(txt, (x[minpnt], y[minpnt]), fontsize = 12,
            xycoords = "data", xytext = (x[int(len(x)/2)],np.max(y)*.6),
            horizontalalignment="center",
            arrowprops = dict(color = "purple", arrowstyle = "wedge"))
plt.show()
```

## Seaborn

```python
# Used particularly in Statistics for advanced graphics and visualization
# it integrates very well with Pandas library
import seaborn as sns
import pandas as pd
```

```python
n = 200
D = np.zeros((n,2))

D[:,0] = np.linspace(0,10,n) + np.random.randn(n)
D[:,1] = D[:,0]**2 + np.random.randn(n)*10

#plt.plot(D[:,0], D[:,1], "k.")
# Joint plot
sns.jointplot(x = D[:,0], y = D[:,1])
plt.show()
```

```python
df = pd.DataFrame(data = D, columns = ["var1", "var2"])
df.head()
```

```python
# kde = kernel smooth estimators: smoothen the plot
# The densities are probability estimations, not empirical distributions
sns.jointplot(x = df.columns[0], y = df.columns[1], data = df,
             kind = "kde", color = "purple", fill = True)
plt.show()
```

```python
# Different way to make scatterplot
x = np.linspace(-1, 1, n)
y1 = x**2
y2 = np.sin(3*x)
y3 = np.exp(-10*x**2) # define colors

# plt.plot(y1,y2)
# palette = jet does not exist
sns.scatterplot(x = y1, y = y2, hue = y3, legend = False, palette="Spectral")
plt.show()
```

```python
# Create a regression plot of the data df above using seaborn
sns.regplot(x = df.columns[0], y = df.columns[1], data = df,
           color = "green")
plt.title(f"Regression of {df.columns[0]} on {df.columns[1]}")
plt.show()
```

```python
# Exercise
# 1) Create a Hilbert matrix (H) (10x10): H[i,j] = 1/(i+j-1)
# 2) Visualize H using seaborn: sns.heatmap()
n = 10
H = np.zeros((n,n))

for i in range(n):
    for j in range(n):
        H[i,j] = 1/(i+j-1+2)
sns.heatmap(H)
```

```python
n = 100
H = np.zeros((n,n))

for i in range(1,n+1):
    for j in range(1,n+1):
        H[i-1,j-1] = 1/(i+j-1)
np.round(H,2)
sns.heatmap(H, vmin = 0, vmax = 0.05)
plt.show()
```

## Serpinski Triangle

```python
n = 10000
x = np.zeros(n)
y = np.zeros(n)
for i in range(1,n):
    k = np.random.choice([1,2,3])
    
    x[i] = 0.5*x[i-1] + k-1
    y[i] = 0.5*y[i-1]
    
    if k == 2:
        y[i] = y[i] + 2
plt.plot(x,y, "k.", markersize = 1)
plt.title("Serpinski Triangle")
plt.axis("off")
plt.show()
```
