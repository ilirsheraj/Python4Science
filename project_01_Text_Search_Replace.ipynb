## Import a Text File

The text file is an autogenerated transcript from an online course. It is a transcript of all words and the time frames they were pronounced. 

We want to strip off all the formatting and time stamp and get only text. Then we will find all 4-letter words and replace them with something else.

```python
import pandas as pd
import numpy as np
import re # regular expression
# pip install webvtt-py
import webvtt
```

```python
file = "/Users/sheraji/OnlineCourses/PythonforScience/MasterPython_CodeAndData/textSearchReplace/captions_text.vtt"
text = webvtt.read(file)
text = str(text)
text
```

```python
sentence = "Hi, my name is Mike and I like to eat Mike's chocolates."

regex_name = re.compile("Mike")
regex_name.findall(sentence)
```

```python
# Substitute
regex_name.sub("MIKE", sentence)
```

```python
regex_name.sub("Mike X", sentence)
```

```python
names = ["Adam", "Bob", "Chrles"]
for name in names:
    print(regex_name.sub(name, sentence))
```

```python
# Wildcard to replace patterns
# Find any word with capital M and anything else, and then "ke"
sentence = "Hi, my name is Mike and I like to eat Muke's chocolates."

regex_name = re.compile(r"M\wke") # \w means any letter
regex_name.findall(sentence)
```

```python
text = text.replace('\n','')
text
```

```python
# Pattern to replace

#\n00:00:05.080 00:00:09.310
                
pattern2replace = r"\d\d:\d\d:\d\d.\d\d\d \d\d:\d\d:\d\d.\d\d\d"
regex_timestamps = re.compile(pattern2replace)
regex_timestamps.findall(text)
```

```python
newtext = regex_timestamps.sub("", text)
newtext = newtext[1:]
newtext
```

```python
# Find all 4-letter words, keep the first letter and replace the other 3 with asterisks
len(newtext)
```

```python
#newtext.split(sep="e")
```

```python
words = newtext.split()
print(len(words))
```

```python
words[:30]
```

```python
#loop over the words and fin all those with 4-lettters
for wordi in words:
    if len(wordi) == 4:
        print(wordi)
```

```python
for i in range(len(words)):
    if len(words[i]) == 4 and not words[i][-1] == ".":
        print(words[i])
```

```python
for i in range(len(words)):
    if len(words[i]) == 4 and not words[i][-1] == ".":
        words[i] = words[i][0] + "***"
cleantext = " ".join(words)
cleantext
```

```python
with open("project1_clean_transcript.txt", "w") as f:
    f.write(cleantext)
```

## Bonus: Readability of Scrambled Words

```python
text = "I have often wondered whether it's possible to read an entire book faster if the useless letters were removed."
text
```

```python
words = text.split()
words
```

```python
# shuffle the numbers from 0-4 randomly
np.random.permutation(5)
```

```python
w = "removed"
idx = np.random.permutation(len(w))
idx
"".join([w[i] for i in idx])
```

```python
# First completely scramble all letters
for j in range(len(words)):
    thisword = words[j]
    idx = np.random.permutation(len(thisword))
    words[j] = "".join([thisword[i] for i in idx])
" ".join(words)
```

```python
# Modify the code above to scrample all letters except first and last
words = text.split()
for j in range(len(words)):
    # ignore short words
    if len(words[j]) < 4:
        continue
        
    thisword = words[j][1:-1]
    idx = np.random.permutation(len(thisword))
    words[j] = words[j][0] + "".join([thisword[i] for i in idx]) \
              + words[j][-1] 
" ".join(words)
```

```python
# replace letters with completely random letters from alphabet
import string
allletters = string.ascii_lowercase
words = text.split()
for j in range(len(words)):
    # ignore short words
    if len(words[j]) < 4:
        continue  
    thisword = words[j][1:-1]
    idx = np.random.randint(0, len(allletters), len(thisword))
    words[j] = words[j][0] + "".join([allletters[i] for i in idx]) + words[j][-1]

" ".join(words)
```
